```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ames = pd.read_csv('./datasets/train_features.csv', index_col = 0)
ames_large = pd.read_csv('./datasets/train_with_dummies.csv', index_col = 0)
```

```python


#sns.set_style('darkgrid')

txtcolor = 'whitesmoke'

sns.set_style({'axes.facecolor' : 'xkcd:cool gray',
              'figure.facecolor' : 'xkcd:slate gray',
              'xtick.color' : txtcolor,
              'ytick.color' : txtcolor,
              'text.color' : txtcolor})


plt.figure(figsize = (16,12))


sns.boxplot(x = ames['MS SubClass'], y = ames['SalePrice'], palette='Spectral')

plt.xlabel('MS SubClass', fontsize = 20, color = txtcolor)
plt.ylabel('Sale Price', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Sale Prices by Sale Sub-Class', fontsize = 26, color = txtcolor)


plt.show()

#sns.axes_style()
```

```python
txtcolor = 'whitesmoke'

sns.set_style({'axes.facecolor' : 'xkcd:cool gray',
              'figure.facecolor' : 'xkcd:slate gray',
              'xtick.color' : txtcolor,
              'ytick.color' : txtcolor,
              'text.color' : txtcolor})


plt.figure(figsize = (16,12))


sns.boxplot(x = ames['Mo Sold'], y = ames['SalePrice'], palette='Spectral')

plt.xlabel('Month', fontsize = 20, color = txtcolor)
plt.ylabel('Sale Price', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Sale Prices by Month', fontsize = 26, color = txtcolor)


plt.show()
```

```python
txtcolor = 'black'

sns.set_style({'axes.grid' : 'True'})


sns.set_style({'axes.facecolor' : 'slategray',
              'figure.facecolor' : 'xkcd:bluegray',
              'xtick.color' : txtcolor,
              'ytick.color' : txtcolor,
              'text.color' : txtcolor})

plt.figure(figsize = (16,12))

#Some code modified from: https://stackoverflow.com/a/35827498
sns.lmplot('Total SF', 'SalePrice', ames, height = 13, ci = None, scatter_kws={'color': 'lightsalmon'}, line_kws={'color': 'sienna'}) 


plt.xlabel('Total Square Feet', fontsize = 20, color = txtcolor)
plt.ylabel('Sale Price', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Sale Prices vs Square Feet', fontsize = 26, color = txtcolor)

plt.show()
```

```python
txtcolor = 'black'

sns.set_style({'axes.facecolor' : 'slategray',
              'figure.facecolor' : 'xkcd:bluegray',
              'xtick.color' : txtcolor,
              'ytick.color' : txtcolor,
              'text.color' : txtcolor})

plt.figure(figsize = (16,12))


sns.lmplot('SF Qual', 'SalePrice', ames, height = 13, ci = None, scatter_kws={'color':'plum'}, line_kws={'color': 'orchid'})


plt.xlabel('Size-Quality Interaction', fontsize = 20, color = txtcolor)
plt.ylabel('Sale Price', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Sale Prices vs Size-Quality Interaction', fontsize = 26, color = txtcolor)

plt.show()
```

```python
resids = ames['SalePrice'] - ames_large['Ridge Preds']
```

```python
txtcolor = 'black'


sns.set_style({'axes.facecolor' : 'slategray',
              'figure.facecolor' : 'xkcd:bluegray',
              'xtick.color' : txtcolor,
              'ytick.color' : txtcolor,
              'text.color' : txtcolor})

plt.figure(figsize = (16,12))
sns.scatterplot(ames['SalePrice'], resids, color = 'xkcd:blue/green')

plt.xlabel('Sale Price', fontsize = 20, color = txtcolor)
plt.ylabel('Error in Prediction', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Error in Model Predictions', fontsize = 26, color = txtcolor)
```

```python
import sklearn.metrics as metrics

metrics.mean_absolute_error(ames['SalePrice'], ames_large['Ridge Preds'])
```

```python
resids = ames['SalePrice'] - ames_large['Ridge Preds']
```

```python
sns.set_style({'axes.grid' : 'True'})

plt.figure(figsize = (16,12))
sns.distplot(resids, label= 'Multiple Linear Regression Model',
              kde_kws={'color':'turquoise'},
              hist_kws={'linewidth':'0', 'color':'xkcd:blue/green', 'alpha' : .8})

sns.distplot(ames['SalePrice'] - ames_large['SF Preds'], label = 'Simple Model',
             hist_kws= {'linewidth':'0', 'alpha' : .5})

plt.xlabel('Error Amount', fontsize = 20, color = txtcolor)
plt.ylabel('Distribution', fontsize = 20, color = txtcolor)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16, rotation = 30)
plt.title('Error in Model Predictions', fontsize = 26, color = txtcolor)

plt.legend(fontsize = 18)

plt.show()
```
